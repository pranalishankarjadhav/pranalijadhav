```python
top=tk.Tk()

top.title('Squre of Number')#for window name
top.geometry('700x500')#for window size

data=tk.IntVar()
def square():
    num=data.get()
    result=num*num
    messagebox.showinfo('MESSAGE','Square is %i'%result)


l1=tk.Label(top,text='INPUT').place(x=50,y=50)
e1=tk.Entry(top,textvariable=data).place(x=100,y=50)
b1=tk.Button(top,text='Squre',command=square,font=('Bold',14),bg='purple',fg='yellow').place(x=120,y=100)
top.mainloop()

```

```python
#take 2 numbers,add,sub,divide,mul

top=tk.Tk()

top.title('Arithmatic Operations')#for window name
top.geometry('700x500')#for window size

data1=tk.IntVar()
data2=tk.IntVar()
def add():
    num1=data1.get()
    num2=data2.get()
    result1=num1+num2
    messagebox.showinfo('MESSAGE','Addition is %i'%result1)
    
def sub():
    num1=data1.get()
    num2=data2.get()
    result2=num1-num2
    messagebox.showinfo('MESSAGE','Substraction is %i'%result2)
    
def mul():
    num1=data1.get()
    num2=data2.get()
    result3=num1*num2
    messagebox.showinfo('MESSAGE','Multiplication is %i'%result3)
def div():
    num1=data1.get()
    num2=data2.get()
    result4=num1/num2
    messagebox.showinfo('MESSAGE','Devision is %i'%result4)
    
    


l1=tk.Label(top,text='INPUT1').place(x=50,y=50)
l2=tk.Label(top,text='INPUT2').place(x=50,y=100)

e1=tk.Entry(top,textvariable=data1).place(x=100,y=50)
e2=tk.Entry(top,textvariable=data2).place(x=100,y=100)

b1=tk.Button(top,text='add',command=add,font=('Bold',14),bg='purple',fg='yellow').place(x=180,y=150)
b2=tk.Button(top,text='sub',command=sub,font=('Bold',14),bg='purple',fg='yellow').place(x=250,y=150)
b3=tk.Button(top,text='mul',command=mul,font=('Bold',14),bg='purple',fg='yellow').place(x=300,y=150)
b4=tk.Button(top,text='div',command=div,font=('Bold',14),bg='purple',fg='yellow').place(x=350,y=150)

top.mainloop()

```
