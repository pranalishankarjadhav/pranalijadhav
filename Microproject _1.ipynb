```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load the dataset


try:
    df = pd.read_csv('Employee Attrition.csv')
except FileNotFoundError:
    messagebox.showerror("Error", "File 'Employee Attrition.csv' not found!")
    exit()
   
        
# Function to get employee information
def get_employee_info():
    employee_id = entry_employee_id.get()
    if not employee_id:
        messagebox.showwarning("Input Error", "Please enter an Employee ID")
        return
    
    try:
        employee_id = int(employee_id)
    except ValueError:
        messagebox.showwarning("Input Error", "Please enter a valid Employee ID")
        return
    
    employee_info = df[df['EmployeeNumber'] == employee_id]
    if employee_info.empty:
        messagebox.showinfo("Not Found", f"No employee found with ID {employee_id}")
    else:
        info_text = ""
        for column in employee_info.columns:
            info_text += f"{column}: {employee_info.iloc[0][column]}\n"
        text_employee_info.delete(1.0, tk.END)
        text_employee_info.insert(tk.END, info_text)

# Create the main Tkinter application window
root = tk.Tk()
root.title('Employee Information')
root.geometry('600x400')
root.configure(bg = 'pink')


# Label and entry for Employee ID
label_employee_id = tk.Label(root, text="Enter Employee ID:", font=('Bold', 14))
label_employee_id.pack(pady=10)
entry_employee_id = tk.Entry(root, font=('Bold', 14))
entry_employee_id.pack(pady=10)

# Button to get employee information
button_get_info = tk.Button(root, text='Get Employee Info', command=get_employee_info, font=('Bold', 14), bg='purple', fg='yellow')
button_get_info.pack(pady=20)


# Text widget to display employee information
text_employee_info = tk.Text(root, font=('Bold', 12), width=60, height=10, wrap='word')
text_employee_info.pack(pady=20)

# Run the Tkinter main loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import Button, messagebox
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Sample DataFrame (replace with your actual data)
data = df['Attrition']
df = pd.DataFrame(data)

# Create the main Tkinter application window
top = tk.Tk()
top.title('Attrition Analysis')
top.geometry('700x500')
top.configure(bg = 'lightblue')


# Function to display count plot using seaborn
def countplot():
    # Clear previous plot (if any)
    for widget in top.winfo_children():
        widget.pack_forget()

    # Create a seaborn countplot
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='Attrition')
    plt.title('Attrition Count')
    plt.xlabel('Attrition')
    plt.ylabel('Count')

    # Embed plot into Tkinter
    canvas = FigureCanvasTkAgg(plt.gcf(), master=top)
    canvas.draw()
    canvas.get_tk_widget().pack()

    # Show the "Back to Menu" button
    back_to_menu_button.pack()

# Function to display pie chart using matplotlib
def pie():
    # Clear previous plot (if any)
    for widget in top.winfo_children():
        widget.pack_forget()

    # Calculate attrition rates
    attrition_count = df['Attrition'].value_counts(normalize=True) * 100

    # Create a pie chart
    plt.figure(figsize=(6, 4))
    plt.pie(attrition_count, labels=attrition_count.index, autopct='%1.1f%%', startangle=140,
            colors=['purple', 'yellow'])
    plt.title('Attrition Rate')

    # Embed plot into Tkinter
    canvas = FigureCanvasTkAgg(plt.gcf(), master=top)
    canvas.draw()
    canvas.get_tk_widget().pack()

    # Show the "Back to Menu" button
    back_to_menu_button.pack()

# Function to go back to main menu
def back_to_menu():
    # Clear current plot
    for widget in top.winfo_children():
        widget.pack_forget()

    # Show the main menu
    show_main_menu()

# Function to display main menu (buttons)
def show_main_menu():
    b1.place(x = 200, y = 200)
    b2.place(x = 400, y = 200)

# Exit function
def exit_application():
    if messagebox.askokcancel("Exit", "Do you want to exit the application?"):
        top.destroy()
        ##The exit_application() function uses messagebox.askokcancel() to confirm exit before destroying the top window.


## Button for display of both graphs
b1 = Button(top, text='Count Plot', command=countplot, font=('Bold', 14), bg='purple', fg='yellow')
b1.place(x = 200, y = 200)


b2 = Button(top, text='Pie Chart', command=pie, font=('Bold', 14), bg='purple', fg='yellow')
b2.place(x = 350, y = 200)

# Exit button
exit_button = tk.Button(top, text="Exit", command=exit_application,font=('Bold', 14))
exit_button.pack(pady=10)

# Back to menu button
back_to_menu_button = tk.Button(top, text="Back to Menu", command=back_to_menu)
back_to_menu_button.pack_forget()  # Initially hide the back to menu button


# Show the main menu initially
show_main_menu()


top.mainloop()
```
